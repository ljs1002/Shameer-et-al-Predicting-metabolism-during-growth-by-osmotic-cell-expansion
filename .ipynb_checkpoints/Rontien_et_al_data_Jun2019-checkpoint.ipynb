```python
Biomass = dict()
Biomass["Starch_b"]=ys_starch(4)*0.01 #x mmol/100mL = 0.01x mmol/mL
Biomass["Cellulose_b"]=ys_cellulose(4)*0.01 #x mmol/100mL = 0.01x mmol/mL
Biomass["sSUCROSE_b"]=ys_sucrose(4)*0.00001 #x umol/100mL = 0.01x umol/mL = (0.01*0.001)x mmol/mL
Biomass["GLC"]=ys_glucose(4)*0.00001 #x umol/100mL = 0.01x umol/mL = (0.01*0.001)x mmol/mL
Biomass["FRU"]=ys_fructose(4)*0.00001 #x umol/100mL = 0.01x umol/mL = (0.01*0.001)x mmol/mL
Biomass["sMAL_b"]=ys_malate(4)*0.00001 #x umol/100mL = 0.01x umol/mL = (0.01*0.001)x mmol/mL
Biomass["sCIT_b"]=ys_citrate(4)*0.00001 #x umol/100mL = 0.01x umol/mL = (0.01*0.001)x mmol/mL
Biomass["Protein_b"]=ys_protein(4)*0.001/398.6940977020116 #x mg/100mL = 0.01x mg/mL = (0.01/398.69)x mmol/mL

#Rontein et al report that lipid flux was negligible, however based on Arabidopsis data (Williams et al 2010) lipid:cellulose content = (11/MW_phospholipid)/(43/MW_cellulose)
Biomass["L_PHOSPHATIDATE_p"]=ys_cellulose(4)*0.01*(11.0/dividing_model.metabolites.get_by_id("L_PHOSPHATIDATE_p").formula_weight)/(43.0/dividing_model.metabolites.get_by_id("CELLULOSE_c").formula_weight)
```

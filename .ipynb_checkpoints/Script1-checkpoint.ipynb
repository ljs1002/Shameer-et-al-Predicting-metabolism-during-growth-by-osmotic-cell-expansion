# Validating equations 3, 4 and 5 used to predict cellulose, phospholipid and protein content in expanding pericarp cells

### Cellulose data

```python
import pandas as pd
d={"DPA":[15,15,15,15,15,15,42,42,42,42,42,42,59,59,59,59,59,59],"CELLULOSE":[334.44,504.83,382.78,417.73,378.01,365.3,4934.22,4600.56,4523.37,4553.37,3943.33,3633.26,3870.56,3891.64,4310.25,4300.31,3338.23,3426.91]}

n=0
for i in d["CELLULOSE"]:         #converting cellulose to mmol/fruit
    d["CELLULOSE"][n]=i/1000
    n=n+1

df_cellulose = pd.DataFrame(data = d)
df_cellulose = df_cellulose[["DPA","CELLULOSE"]]
df_cellulose
```

```python
from studyFunctions import *

xlist = list()
ylist = list()
for i in range(8,61):
    xlist.append(i)
    ylist.append(celluloseLevels(i,Ncells = 25*(10**6)))

SSE = 0
for i in range(0,len(df_cellulose["CELLULOSE"])):
    SSE=SSE+((ylist[df_cellulose["DPA"][i]-8]-df_cellulose["CELLULOSE"][i])**2)
MSE = float(SSE)/len(df_cellulose["DPA"])
RMSE = (MSE)**0.5
RMSE_cellulose = RMSE

```

### Lipid data

```python
import pandas as pd

df_lipid = pd.DataFrame(data={'DPA (days)':[15,15,15,15,15,15,42,42,42,42,42,42,59,59,59,59,59,59],'Lipids(mg/gFW)':[5.66,5.05,3.83,6.15,5.26,4.21,3.60,3.39,2.61,3.75,3.58,2.78,2.94,3.73,3.04,0.70,4.28,3.45]})
FWdict = {8:0.32,15:2.34,22:5.50,28:15.35,34:38.42,42:54.65,50:58.24,52:64.36,59:57.47}
MW_lipid=689.979
temp = list()
for i in range(0,len(df_lipid["Lipids(mg/gFW)"])):
    temp.append((df_lipid["Lipids(mg/gFW)"][i])/MW_lipid)
df_lipid["Lipids (mmol/gFW)"]=temp

temp = list()
for i in range(0,len(df_lipid["Lipids(mg/gFW)"])):
    temp.append(df_lipid["Lipids (mmol/gFW)"][i]*FWdict[df_lipid["DPA (days)"][i]])
df_lipid["Lipids (mmol/fruit)"]=temp
df_lipid
```

```python
from studyFunctions import *

xlist2 = list()
ylist2 = list()
for i in range(8,61):
    xlist2.append(i)
    ylist2.append(phospholipidLevels(i,Ncells = 25*(10**6),scaling_factor=100.0/8.512820512820287))

SSE = 0
for i in range(0,len(df_lipid["DPA (days)"])):
    SSE=SSE+((ylist2[df_lipid["DPA (days)"][i]-8]-df_lipid["Lipids (mmol/fruit)"][i])**2)
MSE = float(SSE)/len(df_lipid["DPA (days)"])
RMSE = (MSE)**0.5
RMSE_lipid = RMSE
```

### Protein data

```python
import pandas as pd
df_protein = pd.read_csv("Data/Data_Starch_Protein_oirginal.csv",sep="\t")
MW = 461.181100469
temp = list()
for i in range(0,len(df_protein["DPA (days)"])):
    temp.append(df_protein["PROTEIN (ug/mgDW)"][i]*df_protein["fruit DW (g)"][i]*1000)
df_protein["PROTEIN (ug/fruit)"]=temp

temp = list()
for i in range(0,len(df_protein["DPA (days)"])):
    temp.append(df_protein["PROTEIN (ug/fruit)"][i]/1000)
df_protein["PROTEIN (mg/fruit)"]=temp

temp = list()
for i in range(0,len(df_protein["DPA (days)"])):
    temp.append(df_protein["PROTEIN (mg/fruit)"][i]/MW)
df_protein["PROTEIN (mmol/fruit)"]=temp

temp = list()
for i in range(0,len(df_protein["DPA (days)"])):
    temp.append(df_protein["PROTEIN (mmol/fruit)"][i]/float(estimateVcyt(df_protein["DPA (days)"][i])))
df_protein["PROTEIN (mmol/m3)"]=temp
df_protein
```

```python
xlist3 = list()
ylist3 = list()
for i in range(8,60):
    xlist3.append(i)
    ylist3.append(estimateProteinContent(i,ProtConc=19659.7704473,Ncell = 25*(10**6)))
    
SSE = 0
for i in range(0,len(df_protein["DPA (days)"])):
    SSE=SSE+((ylist3[df_protein["DPA (days)"][i]-8]-df_protein["PROTEIN (mmol/fruit)"][i])**2)
MSE = float(SSE)/len(df_protein["DPA (days)"])
RMSE = (MSE)**0.5
RMSE_protein = RMSE
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams.update({'font.size': 20}) #sets a global fontsize
plt.rcParams['xtick.major.size'] = 5 # adjusts tick line length and width
plt.rcParams['xtick.major.width'] = 1
plt.rcParams['ytick.major.size'] = 5
plt.rcParams['ytick.major.width'] = 1
plt.rcParams['axes.linewidth']=3 # makes axes line thicker
plt.figure(figsize=(6,15))

ax1 = plt.subplot(3,1,1)
ax1.scatter(df_cellulose["DPA"],df_cellulose["CELLULOSE"],label="Measured",color="black")
ax1.plot(xlist,ylist,"k--",label="Predicted")
#ax1.set_xlabel('DPA', fontsize=16)
ax1.set_ylabel('cellulose content (mmol/fruit)', fontsize=20)
ax1.legend(loc="upper left",fontsize=16)
ax1.text(8,3.8,"RMSD="+str(round(RMSE_cellulose,3)),fontsize=14)

ax2 = plt.subplot(3,1,2)
ax2.scatter(df_lipid["DPA (days)"],df_lipid["Lipids (mmol/fruit)"],label="Measured",color="black")
ax2.plot(xlist2,ylist2,"k--",label="Predicted")
#ax2.set_xlabel('DPA', fontsize=16)
ax2.set_ylabel('lipid content (mmol/fruit)', fontsize=20)
ax2.legend(loc="upper left",fontsize=16)
ax2.text(8,0.28,"RMSD="+str(round(RMSE_lipid,3)),fontsize=14)

ax3 = plt.subplot(3,1,3)
ax3.scatter(df_protein["DPA (days)"],df_protein["PROTEIN (mmol/fruit)"],color="black",label="Measured")
ax3.plot(xlist3,ylist3,"k--",label="Predicted")
ax3.set_xlabel('DPA', fontsize=20)
ax3.set_ylabel("protein content (mmol/fruit)", fontsize=20)
ax3.legend(loc="upper left",fontsize=16)
ax3.text(8,0.155,"RMSD="+str(round(RMSE_protein,3)),fontsize=14)

plt.legend(loc="upper left",fontsize=16)
plt.tight_layout()
plt.show()
```
